```r
library(tidyverse)

full_df <- data.table::fread("./full_df.tsv")
```

```r
full_df[1:5, 1:10]
```

```r
img_arr <- full_df[-40, -c(1:5)]
```

```r
full_df[c(7,40), 1:5]
```

```r
mds <- img_arr %>% 
  dist() %>% 
  MASS::isoMDS() %>%
  #cmdscale() %>% 
  as.data.frame() %>% 
  mutate(method = "mds") %>% 
  #rename(x = V1, y = V2)
  rename(x = `points.1`, y = `points.2`)

pca <- (img_arr %>% prcomp())$x %>% 
  as.data.frame() %>%
  mutate(method = "pca")
```

```r
p1 <- pca %>% 
  ggplot(., aes(PC1, PC2)) + 
  geom_point() +  
  labs(title = "PCA")

p2 <- mds %>% 
  ggplot(., aes(x, y)) + 
  geom_point() +
  labs(title = "MDS")
```

```r
gridExtra::grid.arrange(p1, p2)
```
